```python
usa_commodity_price = pd.read_csv('USA_Commodity_prices.csv')
```

```python
usa_commodity_price
```

```python
train = usa_commodity_price[usa_commodity_price.Year < 2003]
test = usa_commodity_price[usa_commodity_price.Year >= 2003]
```

```python
train.Year = pd.to_datetime(train.Year,format="%Y")
```

```python
train
```

```python
train.index = train.Year
```

```python
train.drop('Year',axis=1,inplace=True)
```

```python
train
```

```python
train['time'] = np.arange(len(train.index))
```

```python
train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-whitegrid")
plt.rc(
    "figure",
    autolayout=True,
    figsize=(11, 4),
    titlesize=18,
    titleweight='bold',
)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=16,
    titlepad=10,
)
%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots()

ax.plot('time', 'Cotton_Price[Dollar/ton]', data=train, color='0.75')
ax = sns.regplot(x='time', y='Cotton_Price[Dollar/ton]', data=train, ci=None, scatter_kws=dict(color='0.25'))
ax.set_title('Time Plot of Cotton_Price[Dollar/ton]');
```

```python
train['Lag_1'] = train['Cotton_Price[Dollar/ton]'].shift(1)
train = train.reindex(columns=['Cotton_Price[Dollar/ton]', 'Lag_1'])
```

```python
fig, ax = plt.subplots()
ax = sns.regplot(x='Lag_1', y='Cotton_Price[Dollar/ton]', data=train, ci=None, scatter_kws=dict(color='0.25'))
ax.set_aspect('equal')
ax.set_title('Lag Plot of Cotton_Price[Dollar/ton]');
```

```python
train.drop('Lag_1',axis=1,inplace=True)
```
